### Function to calculate the final price after GST Tax 

In India, GST(Goods and Services Tax) is calculated by applying the applicable GST rate. The GST rate of most goods and services is 18%. Sales between states, the GST is charged as IGST(Integrated GST).


### Formula:
#### GST Amount = Base Price * GST rate 
#### Final Price = Base Price + GST Amount

```python
def gst_add(base_price):
    # calculating GST amount 
    gst_amount = base_price * (1 - .18)
    # calculating price after GST amount
    price_afterGST = base_price + gst_amount
    return price_afterGST
    
```

```python
final_price = gst_add(20000)
print(final_price)
```

### Function with Multiple arguments such as  discount and GST rate 

#### Formula:
#### Discount Price = Base Price x discont presentage 
#### Price after discount = Base Price - Discount Price
#### GST Amount = Price after discount x GST presentage
#### Price = Price after discount + GST amount 

```python
def gst_with_discount(base_price,discount,gst_rate):
    price_off = base_price * discount / 100
    price_after_discount = base_price - price_off
    gst_amount = price_after_discount * gst_rate/100
    cost_of_items = price_after_discount + gst_amount
    return cost_of_items
```

```python
final_price = gst_with_discount(2000,10,18)
print(final_price)
```
